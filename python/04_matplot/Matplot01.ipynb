파이썬으로 분석한 데이터를 시각화 하는데는 matplotlib 라이브러리를 가장 많이 사용

실제 과학 컴퓨팅 분야나 인공지능 연구분석결과 도출에 가장 많이 사용됨.

```python
import matplotlib.pyplot as plt
```

### Plot 직선 그리기

```python
#plot() : 직선, 혹은 꺽은선 형태의 그래프를 그릴 수 있다. 선형 | 비선형
#간단한 데이터를 하나 만들어서 (x,y) 직선 혹은 꺾은선을 그려보자.
plt.plot([1,2,3,4],[1,4,9,16]) #첫번째 리스트 값 X, 두번째 리스트 값 Y
plt.show()

#2. 직선?
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()

#3. 리스트에 입력한 값은 y축의 값으로 입력되며, X축은 자동으로 0부터 1씩 증가
plt.plot([10,20,30,40])
plt.show()
```

### 겹쳐진 그래프 그리기

```python
import numpy as np
```

```python
t = np.arange(0,5,0.2)
t
```

```python
plt.plot(t,t,'g--')
plt.show()
```

```python
plt.plot(t,t,'r--', t,t**2,'s',t,t**3,'b^') #s는 스퀘어.. ^삼각형, *별표, s--:square점선으로 잇기
plt.show()
```

```python
plt.plot(t,t,'r--',label='Tips')
plt.plot(t,t**2,'s',label='Pop')
plt.plot(t,t**3,'b^',label='Product Layer')
plt.legend() #이 부분을 해야 Label표식이 나타남
plt. title('Wrapped Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid()
plt.show()
```

### 파도모양 그래프

```python
t = np.arange(0,12,0.01)
'''
기본으로 존재하는 Figure가 아닌 새로운 figure생성을 위해서는
plt.figure() 함수를 사용한다.
'''

plt.plot(t, np.sin(t),'r',lw=10,label="sin")
plt.plot(t, np.cos(t),'g',label="cos")

plt.legend()
plt.xlabel('Time')
plt.ylabel('Amplitute')
plt.title('SinCos Graph')
plt.show()
```

### 산점도(Scatter)
- 산점도 그래프는 scatter() 함수를 이용해서 그린다.
- 산점도는 x,y축에 해당하는 데이터들의 상관관계를 표시할 때, 데이터들이 어떻게 흩어져 있는 가를 알기 위해 사용한다.
- 2개의 축을 기준으로 데이터가 얼마나 퍼져있는지(분포도)를 알 수 있기 때문에 산포도 혹은 산점도라고 불린다.

```python
# plt.scatter([1,2,3,4],[10,20,30,40],s=[100,200,300,400],c=['yellow','green','magenta','pink']) #s(ize),c(olor)

plt.scatter([1,2,3,4],[10,20,30,40],s=[100,200,300,400],c=range(4),cmap='jet')
plt.colorbar()
plt.show()
```

```python
plt.scatter('a','b',data=data) #a와,b가 어떻게 분포되어있는지 나옴 label사용 안됨 a,b를 확인할 수 없기 때문
plt.show()
```

```python
plt.scatter('a','b',data=data,c=range(50),cmap='jet')
plt.colorbar()
plt.show()
```

```python
import seaborn as sns
```

```python
cuthalf= (data['a']>25) & (data['b']>30)
data['color']=np.where(cuthalf==True,'Red','Pink')

sns.regplot(x=data['a'],
           y=data['b'],
            fit_reg=False,
           scatter_kws={'facecolors':data['color']})
plt.title('Scatter Plot with Colors', fontsize=20)
plt.show()
```

### Subplot으로 plot(),scatter(),bar() 동시에 이용하기 

```python
names = ['group_A','group_B','group_C'] #x축으로 사용
values = [1,10,100] #Y축으로 사용
```

```python
plt.figure(1,figsize=(9,3))

plt.subplot(131)
plt.bar(names, values)

plt.subplot(132)
plt.scatter(names, values)

plt.subplot(133)
plt.plot(names, values)

plt.show()
```
