### 1. DataLoading and DataPreProcessing 

```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('../data/movies.csv')
ratings = pd.read_csv('../data/ratings_1.csv')
print(movies.shape)
print(ratings.shape)
```

```python
'''
ratings는 사용자별 영화에 대한 평점을 매긴 데이터셋
평점 - 0.5~5.0(0.5단위)
timestamp는 필요 없어서 삭제할 것.
'''
ratings = ratings[['userId','movieId','rating']]

#index=userId, columns = movieId
ratings_matrix = ratings.pivot_table(index='userId',columns='movieId',values='rating')
ratings_matrix
```

```python
'''
영화 ID대신 타이틀이 필요하다.
title 컬럼은 movies에 있다.
title 컬럼을 얻기 위해서는 movies와 조인해야한다. 
'''
rating_movies = pd.merge(ratings,movies,on='movieId')
rating_matrix = rating_movies.pivot_table(index='userId',columns='title',values='rating')

# 2. nan-> 0으로
rating_matrix=rating_matrix.fillna(0)
rating_matrix.head(5)
```

### 2. 영화와 영화들 간의 유사도 산출 - Cosine Similarity 
    - 사용자와 변환된 영화 평점 데이터 행렬 데이터 셋을 사용하여 영화간 유사도를 측정
    - 코사인 유사도는 기준이 되는 행과 다른 행을 비교해서 유사도를 특정하는 함수.
    - 지금과 같은 행렬구조는 userId가 기준인 행 레벨 데이터이기 때문에 사용자간의 유사도가 만들어진다. 이것을 사용자가 아닌 
    영화와 영화들간의 유사도를 만들려면 어떻게 해야하는가?
    
    참고로 코사인 유사도 라이브러리는 사이킷런의 cosine_similarity()함수를 사용한다.

```python
rating_matrix_T = rating_matrix.transpose()
rating_matrix_T.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
item_sim = cosine_similarity(rating_matrix_T,rating_matrix_T)
item_sim_df = pd.DataFrame(item_sim,index=rating_matrix.columns, columns=rating_matrix.columns)
item_sim_df
```

```python
item_sim_df['Godfather, The (1972)'].sort_values(ascending=False).head(6)
```

```python
item_sim_df['Inception (2010)'].sort_values(ascending=False)[1:5]
```

##### Case 1.
    - 입력한 영화에 대해 유사도가 제일 높은 영화들을 추천

```python
def high_scoring(movie_name):
    movieDf = item_sim_df.loc[movie_name].drop(movie_name)
    movieDf2 = movieDf[item_sim_df.loc[movie_name] == movieDf.max()]
    
    return movieDf2.index
```

```python
high_scoring("Dark Knight, The (2008)")
```

##### Case 2.
    -고객이 Case1에서 추천한 영화를 보았다고 가정할때, 그 다음으로 띄워줄 영화의 리스트 생성
    (입력한 영화와, 추천한 영화 모두 제외하고 추천하며 유사도가 0.6 이상 있을 경우만 추천)

```python
def second_recommendation(movie_name):
    title_array = np.array(high_scoring(movie_name))
    title_array = np.append(title_array,[movie_name])

    recommendation_df = item_sim_df[high_scoring(movie_name)].drop(title_array)
    return recommendation_df[recommendation_df>=0.6].dropna().index
```

```python
second_recommendation("Dark Knight, The (2008)")
```

##### Case 3.
    - 고객의 선호도를 바탕으로 보지않은 영화 5개 추천 (고객간 선호도가 0.1가 넘을때 비슷하다고 가정)

```python
#고객간 유사도 계산
user_sim = cosine_similarity(rating_matrix,rating_matrix)
user_sim_df = pd.DataFrame(user_sim,index=rating_matrix.index, columns=rating_matrix.index)
user_sim_df
```

```python
def find_best_match(user_id):
    user_info = user_sim_df.loc[user_id].drop(user_id)
    best_match = user_info[user_info == user_info.max()]
    return rating_matrix_T[best_match.index].sort_values(by=(best_match.index[0]),ascending=False).head(5).index
```

```python
find_best_match(1)
```
