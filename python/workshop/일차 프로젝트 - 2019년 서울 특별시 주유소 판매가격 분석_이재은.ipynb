## 3. 데이터 가공 및 분석

```python
df.info()
```

```python
df['상표'].unique()
```

```python
df['지역'].unique()
```

```python
area=df['지역'].str.split(" ")
df['시'] = area.str.get(0)
df['구'] = area.str.get(1)
df
```

```python
df['시']='서울특별시'
df
```

```python
#파이썬에서 int는 초로 인식하기 때문에 문자열로 변환시켜주어야함

df['기간'] = pd.to_datetime(df['기간'].astype(str), format='%Y-%m-%d')
df
```

```python
df['년']=df['기간'].dt.year
df['월']=df['기간'].dt.month
df['일']=df['기간'].dt.day
df['요일']=df['기간'].dt.day_name() #weekday_name안됨
df
```

```python
df.sort_values(by='휘발유',ascending=False).head(5)
```

```python
df[df['휘발유']!=0].sort_values(by='휘발유').head(5)
```

```python
brand = df.pivot_table(values='휘발유',index='상표')
brand
```

```python
# brand.plot(kind='barh',xlim=(1400,1625))
# plt.show()

brandBar = sns.barplot(x=brand['휘발유'], y=brand.index, palette="vlag")
brandBar.set(xlim=(1400,1625))
```

```python
selfPrice = df.pivot_table(values='휘발유',index=['상표','셀프여부'])
selfPrice
```

```python
s=sns.barplot(x=df['휘발유'], y=df['상표'],hue=df['셀프여부'],color='r')
s.set(xlim=(1400,1700))
```

```python
# 셀프여부에 따라 휘발유 가격을 boxplot으로 시각화해보세요.
# 휘발유가 0인 데이터는 제외하고 시각화합니다.
df2 = df[df['휘발유']!=0]
sns.boxplot(x='셀프여부',y='휘발유',data=df2)
```

```python
# 휘발유 가격과 일반 주유소의 가격을 boxplot을 이용하여 시각화해보세요.
# 휘발유가 0인 데이터는 제외하고 시각화합니다.
plt.figure(figsize=(20,10))
sns.boxplot(x='상표',y='휘발유',hue='셀프여부',data=df2)
```

```python
district = df.pivot_table(values='휘발유',index='구').sort_values(by='휘발유',ascending=False)
district
```

```python
# 지역별 평균 휘발유 가격을 시각화하세요.
district.plot(kind='barh',figsize=(12,8),xlim=(1425,2000))
```

```python
month=df.pivot_table(values='휘발유',index='월')
```

```python
# 월별 평균 휘발유 가격을 시각화 하세요.
month.plot()
plt.xscale
```

```python
diMonth=df.pivot_table(values='휘발유',index='월',columns='구')
diMonth
```

```python
# pivot_table 함수를 이용하여 구와 월 컬럼을 이용하여 평균 휘발유 가격을 시각화 하세요 (heatmap 이용)
plt.figure(figsize=(40,20))
sns.heatmap(diMonth,annot=True, fmt= 'f')
plt.show()
```
